```python
#Indexing, Selecting and Assigning:
```

```python
reviews = pd.read_csv("winemag-data-130k-v2.csv", index_col=0)
```

```python
reviews.country #Accessing country property of reviews.
```

```python
reviews['country'] #another way to access country property.
# both methods can access a series but dot notation won't work for reserved characters.
# use dot notations with simple column name.
```

```python
reviews['country'][0]
```

```python
reviews.iloc[0] #iloc and loc are pandas accessor operator.
```

```python
reviews.iloc[:, 0] # ':' means everything
```

```python
reviews.iloc[:3, 0] # selecting 1st, 2nd, 3rd rows of country column
```

```python
reviews.iloc[1:3, 0] # selecting just 2nd and 3rd rows.
```

```python
reviews.iloc[[0,1,2], 0] # passing a list.
```

```python
reviews.iloc[-5:] # -ve no. can be used for selection.
# they start from end.
```

```python
#loc operator(label-based selection)
reviews.loc[0, 'country']
```

```python
reviews.loc[:, ['taster_name', 'taster_twitter_handle', 'points']]
#end index is inclusive in loc and exclusive in iloc.
```
